```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0,5,11)

y = x**2
```

```python
x
```

```python
y
```

## Functional Method to plot

```python
plt.plot(x,y)
#Notice here we get plot in out cell
```

```python
#And here we'll print plot on screen. Need to use plt.show if not using jupyter notebook
plt.plot(x,y)
plt.show()
```

```python
#OK let's add labels to graph
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#plt.subplot(# of rows, # of cols, Plot number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Now let's see object oriented way
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
#axes is a object of multiple axes
print(axes)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Title')

axes[1].plot(y,x)
axes[1].set_title('Second Title')
plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(5,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,4), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#Save the figure
fig.savefig('my_picture.png', dpi = 200)
```

```python
#Adding legends
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'X Squared')
axes.plot(x,x**3, label = 'X Cubed')
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='green',linewidth=0.7,alpha=0.8,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='green',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')
axes.set_xlim([0,2])
axes.set_ylim([0,2])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='green',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')
axes.axis('tight') #best fits the range on both axis
```

```python
#Using logrethmic axis
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, x, np.exp(x))
axes[1].plot(x,y, x, np.exp(x))
axes[1].set_yscale('log')
```
