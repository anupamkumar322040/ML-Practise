```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='sex',col='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='day',col='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='time',col='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='day',col='smoker',hue='sex',aspect=1.2,height=4)
```
